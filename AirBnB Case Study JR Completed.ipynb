2. How many neighborhood groups are available and which shows up the most?


```python
air_bnb.groupby('neighbourhood_group', as_index=True).neighbourhood_group.count()
#There are 5 neighbourhood groups, Manhattan has the most records. 
```

3. Are private rooms the most popular in manhattan?



```python
air_bnb.groupby(['neighbourhood_group', 'room_type']).neighbourhood_group.count().sort_values(ascending=False)
#Anser: the most popular room_type in Manhattan is Entire House/APT. Private room is second most popular at 7982
```

#4. Which hosts are the busiest based on their reviews?


```python
air_bnb.groupby('host_id').number_of_reviews.sum().sort_values(ascending=False)
#Answer: The host with the most reviews is ID 2273 with 2273 reviews. 
```

5. Which neighorhood group has the highest average price?


```python
air_bnb.groupby('neighbourhood_group').price.mean().sort_values(ascending=False)
#Manhattan has the highest average price at $196.88
```

6. Which neighborhood group has the highest total price?


```python
air_bnb.groupby('neighbourhood_group').price.sum().sort_values(ascending=False).nlargest(5)
#Manhattan has the highest total price $4,264,527
```

7. Which top 5 hosts have the highest total price?


```python
air_bnb.groupby('host_id').sum()[['number_of_reviews','price']].sort_values(by='number_of_reviews', ascending=False)
#Answer, top 5 hosts based on number of reviews are shown below. 
#Host ID 35524316 has the highest total price at $2,055
```
